```python
import random 
def criar_Lista():
    lista= []
    i= 0 
    while i<10:
        num =random.randint(1, 100) 
        i= i+1
        lista.append(num)
    return lista


def ler_Lista(lista):
    
    print("lista", lista) 
    return lista


   
def soma(lista):
    soma= 0 
    for num in lista:
        soma= soma+ num 
    return soma 




def media_lista(lista):
    return soma(lista)/ 10




def menor(lista):
    menor= 101
    for num in lista:
        if num < menor:
            menor = num 
    return menor 


def maior(lista):
    maior= 0
    for num in lista:
        if num > maior:
            maior = num 
    return maior


def esta_ordenadacres(lista):
    i=1
    
    while i<len(lista):
        if lista[i] < lista [i-1]:

            return False
        i= i+1
    return True




def esta_ordenadadecr(lista):
    i=1
    while i<len(lista):
        if lista[i]> lista[i-1]:
            return False
        
        
        i= i+1 
    return True

def procura(lista, elem):
    
    for elem in lista:
        cond = True
        if elem in lista and cond== True:
            print("Esta")
            return False
        else:
            print ("nao esta ")
    


def mostrar_menu():
    print("\nMENU:")
    print("(1) Criar Lista")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) estaOrdenada por ordem crescente")
    print("(8) estaOrdenada por ordem decrescente")
    print("(9) Procura um elemento")
    print("(0) Sair")


lista = []
mostrar_menu()
cond = True 
while cond: 
    
    opcao = (input("Escolha uma opção: "))

    if opcao == '1':
        lista = criar_Lista()
        print("Lista aleatória criada:", lista)

    elif opcao == '2':
        lista = ler_Lista(lista)
        

    elif opcao == '3':
        print("Soma dos elementos:", soma(lista))

    elif opcao == '4':
        print("Média dos elementos:", media_lista(lista))

    elif opcao == '5':
        maior_num = maior(lista)
        print(f"Maior elemento:{maior_num}")

    elif opcao == '6':
        menor_num = menor(lista)
        print(f"Menor elemento:{menor_num}")

    elif opcao == '7':
        if esta_ordenadacres(lista):
         print(" Está ordenada de forma crescente")
        else:
            print(" Não está ordenada de forma crescente ")
            
            

    elif opcao == '8':
        if esta_ordenadadecr(lista):
            print(" A lista está ordenada de forma decrescente ")
        else:
            print(" A lista não está ordenada de forma decresente")

    elif opcao == '9':
        
        elem = int(input("Introduza o número que deseja procurar: "))
        procura(lista, elem)

    elif opcao == '0':
       
        print("Saindo... Lista atual:", lista)
        cond= False
    else:
        print("opção invalida tenta novamente")
        
```
