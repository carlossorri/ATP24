```python
import random

def jogo_fosforos():
    print("Bem-vindo ao jogo dos 21 fósforos!")
    print("1: Jogador começa")
    print("2: Computador começa")
    escolha = input("Escolha o modo de jogo (1 ou 2): ")

    fosforos = 21
    jogo_ativo = True 

    
    def jogada_jogador():
        jogada = 0
        while jogada not in [1, 2, 3, 4]:
            jogada = input("Quantos fósforos quer tirar (1, 2, 3 ou 4)? ")
            if jogada.isdigit():
                jogada = int(jogada)
                if  jogada not in [1, 2, 3, 4]:
                    print("Escolha inválida. Tem de tirar entre 1 e 4 fósforos.")
                    jogada = 0 
            else: 
                print (" Entrada invalida")
           
        return jogada

    # Função para a jogada do computador
    def jogada_computador(fosforos_restantes):
        # Estratégia: Deixar sempre um múltiplo de 5 para o jogador se possível
        jogada = (fosforos_restantes - 1) % 5
        if jogada == 0 or jogada > 4:
            # Se não for possível deixar múltiplo de 5, faz uma jogada aleatória válida
            jogada = random.randint(1, 4)
        print(f"Computador tira {jogada} fósforo(s).")
        return jogada

    # Modo 1: Jogador começa
    if escolha == '1':
        while jogo_ativo:
            # Turno do jogador
            print(f"\n estao disponiveis:{fosforos} fosforos ")
            jogada = jogada_jogador()
            fosforos = fosforos - jogada
            if fosforos <= 0:
                print("O jogador tirou o último fósforo! O jogador perdeu!")
                jogo_ativo = False
                
            print(f"\nFósforos restantes: {fosforos}")
            jogada = jogada_computador(fosforos)
            fosforos -= jogada
            if fosforos <= 0:
                print("O computador tirou o último fósforo! O computador perdeu!")
                jogo_ativo = False

    # Modo 2: Computador começa
    elif escolha == '2':
        while jogo_ativo:
            # Turno do computador
            print(f"\n estao disponiveis:{fosforos} fosforos ")
            jogada = jogada_computador(fosforos)
            fosforos = fosforos -jogada
            if fosforos <= 0:
                print("O computador tirou o último fósforo! O computador perdeu!")
                jogo_ativo = False
            else:
                print(f"\nFósforos restantes: {fosforos}")
                jogada = jogada_jogador()
                fosforos = fosforos- jogada
                if fosforos <= 0:
                    print("O jogador tirou o último fósforo! O jogador perdeu!")
                    jogo_ativo = False
                
# Executar o jogo
jogo_fosforos()
```
